## Q1

```python
from tabulate import tabulate
list1 = []
def ins(id,name,dept,yos):
    tuples = (id,name,dept,yos)
    list1.append(tuples)
    return list1

def update(id,uyos):
    for i in range(len(list1)):
        if id in list1[i]:
            l = list(list1[i])
            l[3] = uyos
            list1[i] = tuple(l)
            break
    else:
        print(f"{id} not found")
    return list1

def delete(id):
    for i in range(len(list1)):
        if id in list1[i]:
            print(f"{list1[i]} has been deleted")
            del list1[i]
            break
    else:
        print(f"{id} not found")
    return list1

def show():
    headers = ["ID", "Name", "Department", "Year of Service"]
    return tabulate(list1,headers=headers,tablefmt="grid")



while(True):
    print("1) Enter a Record\n2) Update a Record\n3) Delete a Record\n4) Show \n5) Exit")
    inp = int(input("Enter a number: "))
    match inp:
        case 1:
            id = int(input("Enter the ID: "))
            name = input("Enter the Name: ")
            dept = input("Enter the Department: ")
            yos = int(input("Enter the Year of Service: "))
            ins(id,name,dept,yos)
        case 2:
            id = int(input("Enter the ID: "))
            uyos = int(input("Enter the Updated year of Service: "))
            update(id,uyos)
        case 3:
            id = int(input("Enter the ID: "))
            delete(id)
        case 4:
            print(show())
        case 5:
            print("Exiting program...")
            break
        case _:
            print("Invalid input")
```

## Q2

```python
import string

def preprocess_review(review):
    review = review.lower()
    review = review.translate(str.maketrans('', '', string.punctuation))  
    review = ''.join([i for i in review if not i.isdigit()])
    return review

def insert_review(collection, review):
    collection.append(preprocess_review(review))

def delete_review(collection, review):
    if review in collection:
        collection.remove(review)
        print("Review deleted successfully.")
    else:
        print("Review not found.")

def update_review(collection, old_review, new_review):
    if old_review in collection:
        index = collection.index(old_review)
        collection[index] = new_review
        print("Review updated successfully.")
    else:
        print("Review not found.")

def print_reviews(collection):
    print("All reviews:")
    for review in collection:
        print(review)


reviews = [] 
while(True):
    print("1) Preprocess a review\n2) Insert a new review\n3) Delete a review\n4) Update a review \n5) Print all reviews\n6) Exit")
    inp = int(input("Enter a number: "))
    match inp:
        case 1:
            review = input("Enter the review: ")
            preprocessed_review = preprocess_review(review)
            print("Preprocessed review:", preprocessed_review)
        case 2:
            review = input("Enter the review: ")
            insert_review(reviews, review)
            print("Review inserted successfully.")
        case 3:
            review = input("Enter the review to delete: ")
            delete_review(reviews, review)
        case 4:
            old_review = input("Enter the review to update: ")
            new_review = input("Enter the new review: ")
            update_review(reviews, old_review, new_review)
        case 5:
            print_reviews(reviews)
        case 6:
            print("Exiting program...")
            break
        case _:
            print("Invalid input")
```
