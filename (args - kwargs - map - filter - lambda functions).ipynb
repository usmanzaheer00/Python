```python
def insert_student_info(student_list, **kwargs):
    student_list.append(kwargs)

def display_student_info(**kwargs):
    print("Name:", kwargs['name'])
    print("Age:", kwargs['age'])
    print("Test Scores:", kwargs['test_scores'])

def calculate_average_scores(*student_list):
    def average(scores):
        return sum(scores) / len(scores)
    
    return list(map(lambda student: (student['name'], average(student['test_scores'])), student_list))

def filter_students_by_age(student_list, min_age):
    return list(filter(lambda student: student['age'] < min_age, student_list))

def display_filtered_students(filtered_students):
    print("Filtered Students:")
    for student in filtered_students:
        display_student_info(**student)

def adjust_test_scores(student_list, adjustment_func, *args):
    for student in student_list:
        student['test_scores'] = adjustment_func(student['test_scores'], *args)

def display_adjusted_students(student_list):
    print("Adjusted Students:")
    for student in student_list:
        display_student_info(**student)

students = [
    {'name': 'Alice', 'age': 20, 'test_scores': [85, 90, 95]},
    {'name': 'Bob', 'age': 22, 'test_scores': [75, 80, 85]},
    {'name': 'Charlie', 'age': 18, 'test_scores': [90, 92, 88]},
    {'name': 'David', 'age': 21, 'test_scores': [78, 85, 80]},
]

insert_student_info(students, name='Eve', age=19, test_scores=[88, 90, 92])
print("Displaying Student Information:")
for student in students:
    display_student_info(**student)

average_scores = calculate_average_scores(*students)
print("\nAverage Test Scores:")
for name, avg_score in average_scores:
    print(f"{name}: {avg_score}")

filtered_students = filter_students_by_age(students, 20)
display_filtered_students(filtered_students)

adjust_test_scores(students, lambda scores, adjustment: [score + adjustment for score in scores], 5)
display_adjusted_students(students)
```
